```python
import math
import json
import itertools
import numpy
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot
from plotly.graph_objs import *

init_notebook_mode(connected=True)

events = json.load(open('data/clean_data.json'))['data']
events = sorted(events, key=lambda e: e['date'])
```

#### Predictions
Games are selected by choosing the lower payout.  We score each game using the difference between the higher and lower payout, and then perform hard filtering to exclude any blacklisted games.

The top scoring predictions should be selected and betted on

```python
BLACK_LISTED_SPORTS = []
def predict_game(game):
    score = 0
    score += math.fabs(game['h'] - game['v'])
    if game['sport'] in BLACK_LISTED_SPORTS or len(game['outcomes']) == 0:
        score -= 10

    prediction = 'h' if game['h'] < game['v'] else 'v'
    payout = min(game['h'], game['v']) if prediction in game['outcomes'] else -1

    return { 'prediction': prediction, 'score' : score, 'payout' : payout, 'game': game }

def predict_new_game(game):
    score = 0
    score += math.fabs(game['h'] - game['v'])
    if game['sport'] in BLACK_LISTED_SPORTS:
        score -= 10

    prediction = 'h' if game['h'] < game['v'] else 'v'
    payout = min(game['h'], game['v']) if prediction in game['outcomes'] else -1

    return { 'prediction': prediction, 'score' : score, 'payout' : payout, 'game': game }
```

#### Evaluating the prediction method
The most optimal prediction method yeilds higher scores for positive payouts 

```python
predictions = map((lambda e: map(predict_game, e['games'])), events)
predictions = list(itertools.chain.from_iterable(predictions))

x = map((lambda p: p['score']), predictions)
y = map((lambda p: p['payout']), predictions)

data = [Scatter(
            x=x,
            y=y,
            mode='markers'
        )]

iplot(data, filename='jupyter/basic_bar')
```

```python
def select_games(games):
    predictions = map(predict_game, games)
    sorted_predictions = sorted(predictions, key=lambda p: p['score'], reverse=True)
    return sorted_predictions[0:3]

def select_new_games(games):
    predictions = map(predict_new_game, games)
    sorted_predictions = sorted(predictions, key=lambda p: p['score'], reverse=True)
    return sorted_predictions[0:3]

def build_ticket(games):
    selected_games = select_games(games)
    payouts = map((lambda g: g['payout']), selected_games)
    return -1 if -1 in payouts else sum(payouts)
```

```python

x = map((lambda e: e['date']), events)
y = map((lambda e: build_ticket(e['games'])), events)

data = [Scatter(
            x=x,
            y=numpy.cumsum(y),
            mode = 'lines+markers'
        )]

iplot(data, filename='jupyter/basic_bar')
```
